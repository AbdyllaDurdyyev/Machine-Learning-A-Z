# Split Datasets

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Setting Seed values
np.random.seed(0)
```

```python
#Creating 30 people dataset with random height and weight
```

```python
height = np.random.normal(180, 10, 30)
weight = np.random.normal(80, 15, 30)
```

```python
#Create pandas dataFrame
data = pd.DataFrame({"Height":height, "Weight":weight})
```

```python
#Print datas
print("Dataset:\n", data.head())
```

```python
#separate the data into independent (X) and dependent (y) variables.
x = data[["Height"]]
y = data[["Weight"]]
```

```python
#Separate data into educational and test sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

```python
#Print education and test sets' dimentions
print("\nEducation set dimention:", X_train.shape)
print("Test set dimrntion:", X_test.shape)
```
